```python
# First fetch the content of the webpage

import requests

url = "https://books.toscrape.com/"
response = requests.get(url)
response
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())  # This formats the HTML in a readable way
```

```python
#Function to extract the ratings of the different books
def get_rating(star_class):
    rating_map={"One":1, "Two":2, "Three":3, "Four":4, "Five":5}
    for key in rating_map:
        if key in star_class:
            return rating_map[key]
    return 0
```

```python
#Fuction to extract the price of the different books
import re  # For regular expression handling

def get_price(price_str):
    # Remove all non-numeric characters except the decimal point
    price_str = re.sub(r'[^0-9.]', '', price_str)
    return float(price_str) if price_str else 0.0  # Convert to float, or 0.0 if empty
```

```python
import pandas as pd
#Fuction to extract only data with this parameters: rating above 4 and price below 20
def scrape_books(filters=None):
    base_url = "https://books.toscrape.com/catalogue/page-{}.html"
    book_data = []
    page=1
    
    # Filters applied to the scraping according the exercise are the seeting ones, but can be changed if desired 
    if filters is None:
        filters = {}
    
    min_rating = filters.get("min_rating", 4)  # Minimum rating (default is 4)
    max_price = filters.get("max_price", 20)   # Maximum price (default is 20)
    min_price = filters.get("min_price", 0)    # Minimum price (default is 0)

    while True:
        print(f"Scraping page {page}...")  # Print message when scraping each page

        response = requests.get(base_url.format(page))
        if response.status_code != 200:
            print(f"Not page {page} found , ending scraping.")
            break
        
        soup = BeautifulSoup(response.text, "html.parser")
        books = soup.find_all("article", class_="product_pod")
        
        if not books:
            print("No more books, finalising.")
            break
        
        for book in books:
            title = book.h3.a["title"]
            star_class = book.p["class"]
            rating = get_rating(star_class)
            
            price_str = book.find("p", class_="price_color").text.strip()
            price = get_price(price_str)
            
            availability = book.find("p", class_="instock availability").text.strip()
            
            detail_url = "https://books.toscrape.com/catalogue/" + book.h3.a["href"]
            detail_response = requests.get(detail_url)
            detail_soup = BeautifulSoup(detail_response.text, "html.parser")
            
            upc = detail_soup.find("th", text="UPC").find_next_sibling("td").text.strip()
            genre = detail_soup.find("ul", class_="breadcrumb").find_all("li")[2].text.strip()
            
            description_tag = detail_soup.find("meta", attrs={"name": "description"})
            description = description_tag["content"].strip() if description_tag else "No description available"
            
            if rating >= min_rating and price <= max_price:
                book_data.append([upc, title, price, rating, genre, availability, description])
        
        page += 1
    
    df_books = pd.DataFrame(book_data, columns=["UPC", "Title", "Price (£)", "Rating", "Genre", "Availability", "Description"])
    return df_books
```

```python
# Define the filters if desired to be differet to the default ones (min_rating=4, max_price=20, min_price=0)
#filters = {
    #"min_rating": 4,    # Minimum rating of 4 stars
    #"max_price": 20      # Maximum price of £5
    #"min_price": 0      # Minimum price of £0
#}

# Call the scrape_books function with the filters
df_books = scrape_books() #Add the filters variable if desired to be different to the default Example of fuction: df_books=scrape_books(filters)

# Display the filtered books
print(df_books)
```
